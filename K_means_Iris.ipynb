```
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```
df = pd.read_csv('https://raw.githubusercontent.com/sandeco/CanalSandeco/master/K-means/iris.csv')
```

```
df
```

```
sb.pairplot(df,hue="target")
```

```
X = np.array(df.drop('target', axis = 1)) #drop  exclui essa coluna 'target'
X
```

# **K-means**

```
from sklearn.cluster import KMeans
```

```
kmeans = KMeans(n_clusters=3, random_state=0)
```

```
kmeans.fit(X)
```

```
kmeans.labels_
```

#**Comparando as classes criadas com as classes originais**

```
df['K-classes'] = kmeans.labels_
df
```

```
sb.pairplot(df,hue='target')
```

```
sb.pairplot(df,hue='K-classes')
```

## **Fazendo um Dendograma**

```
import  plotly.figure_factory  as  ff 
```

```
Y = np.array(df.drop(columns=['target', 'K-classes']))
Y
```

```
fig  =  ff . create_dendrogram (Y) 
fig.update_layout(width=800, height=500)
fig.show()
```
